<h1>Gruping Data</h1>

<h2>K-Means</h2>

```python
distortions = []
K = range(1,20)
for k in K:
    kmeansModel = KMeans(n_clusters=k)
    kmeansModel.fit_predict(PCA_Data)
    distortions.append(kmeansModel.inertia_)
    
plt.figure(figsize=(8,4))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Using the Elbow Method we can choose $k=4$ or $k=5$

We choose $k=5$

```python
Nclusters=5
kmeansModel = KMeans(n_clusters=Nclusters)
y_KMeans = kmeansModel.fit_predict(PCA_Data)

KMDF=PCA_Data
KMDF['CLUSTER ID']=y_KMeans
print(KMDF.head())

print('Number of data in each group ')
print('C0= '+str(sum(y_KMeans==0)))
print('C1= '+str(sum(y_KMeans==1)))
print('C2= '+str(sum(y_KMeans==2)))
print('C3= '+str(sum(y_KMeans==3)))
print('C4= '+str(sum(y_KMeans==4)))
```

```python
plotData2D(PCA_Data,y_KMeans,names=['PC1', 'PC2'],Ntargets=Nclusters)
plotData3D(PCA_Data,y_KMeans,Ntargets=Nclusters)
```
